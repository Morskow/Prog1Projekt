## Osnovne ugotovitve

Najprej navedimo prvih 20 tekmovalcev.

```python
podatki.head(20)
```

Zanima nas, kakšna je porazdelitev najboljših 20 tekmovalcev po državah.

```python
drzave=podatki.head(20).groupby('drzava')
drzave.size()
```

Opazimo, da imajo Kitajska, Koreja in Združene države Amerike očitno veliko več tekmovalcev na samem vrhu, ostale države pa kvečjemu enega.

Zanima nas tudi, kdo vse so bili ljudje, ki so pravilno rešili vse naloge prvega dne.

```python
podatki[(podatki.naloga1==7) & (podatki.naloga2==7) & (podatki.naloga3==7)]
```

Vidimo, da so skoraj vsi tekmovalci, ki so pravilno rešili vse tri naloge prvega dne, rešili tudi prvi dve nalogi drugega dne in dobili zlato medaljo. Ker vidimo, da je peta naloga v tej množici ljudi celo bolje reševana od četrte, si poglejmo povprečno vrednost rezultata vseh tekmovalcev na posameznih nalogah.

```python
podatki[['naloga1', 'naloga2', 'naloga3', 'naloga4', 'naloga5', 'naloga6']].mean()
```

Podkrepili smo naše ugibanje, da je bila 5. naloga podobne težavnosti kot 4. Vidimo še, da je bila 1. naloga konkretno boljše reševana od vseh ostalih.

Seveda nas zanima tudi, kako se je odrezala slovenska ekipa.

```python
podatki[podatki.drzava=='Slovenija']
```

Izvedeli bi radi še, kakšne so bile meje za medalje.

```python
zlato=podatki[podatki.nagrada=='Zlata medalja'].iloc[-1]['skupaj']
srebro=podatki[podatki.nagrada=='Srebrna medalja'].iloc[-1]['skupaj']
bron=podatki[podatki.nagrada=='Bronasta medalja'].iloc[-1]['skupaj']

meje={
    'medalje':["Zlata medalja", "Srebrna medalja", "Bronasta medalja"],
    'meje':[zlato, srebro, bron]
}
pd.DataFrame(meje)
```

Vidimo, da je bilo za vsako od medalj dovolj, da smo rešili dve, tri ali štiri naloge ter dobili še 3 točke od ostalih.
## Četrta naloga
Zanima nas nekaj podrobnejših dejstev o četrti nalogi na tekmovanju. Najprej bi radi narisali graf, ki predstavlja število tekmovalcev, ki so dosegli posamezno število točk na tej nalogi.

```python
podatki.groupby('naloga4').size().plot.bar()
```

Ker je četrta naloga običajno med najlažjima dvema na tekmovanju in lahko vidimo, da je veliko tekmovalcev ni rešilo, nas zanima še, kako izgledajo točke na njej v odvisnosti od skupnih točk tekmovalca.

```python
podatki.plot.scatter('naloga4', 'skupaj')
```

Ker je skupna vsota točk deloma odvisna tudi od števila točk na četrti nalogi, seveda graf ni toliko zgovoren, vseeno pa lahko opazimo, denimo, da je relativno zelo malo oseb, ki so dobile manj kot 4 točke na tej nalogi in so osvojile medaljo (če upoštevamo zgoraj izračunano mejo 17 za bronasto medaljo).

```python
podatki[podatki.naloga4<4].groupby('nagrada').size()
```

Pogledali smo si še, koliko ljudi je dobilo posamezno nagrado in obenem doseglo manj kot 4 točke na tej nalogi, torej smo potrdili opazko, ki smo jo razbrali z grafa.
